```python
#-----------------------------------------------------------------------------
#      PRE-PROCESSAMENTO DE UMA RUN SÓ PARA VISUALIZAÇÃO DO RESULTADO
#-----------------------------------------------------------------------------

import mne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================================================================
# CONFIGURAÇÕES INICIAIS
# =========================================================================
arquivo = "/content/drive/MyDrive/IA753-Trabalho_Final/Data/S02_ME/motorexecution_subject2_run2.gdf"
CANAIS_MOTOR = ["C3", "C1", "Cz", "C2", "C4"]

# =========================================================================
# STEP 1: CARREGAMENTO E LIMPEZA DE INTRUSOS
# =========================================================================
print("Carregando arquivo GDF...")
raw = mne.io.read_raw_gdf(arquivo, preload=True, verbose=False)

canais_remover = [
    'thumb_near', 'thumb_far', 'thumb_index', 'index_near', 'index_far',
    'index_middle', 'middle_near', 'middle_far', 'middle_ring', 'ring_near',
    'ring_far', 'ring_little', 'litte_near', 'litte_far', 'thumb_palm',
    'wrist_bend', 'roll', 'pitch', 'gesture', 'handPosX', 'handPosY',
    'handPosZ', 'elbowPosX', 'elbowPosY', 'elbowPosZ', 'ShoulderAdductio',
    'ShoulderFlexionE', 'ShoulderRotation', 'Elbow', 'ProSupination',
    'Wrist', 'GripPressure'
]
raw.drop_channels([ch for ch in canais_remover if ch in raw.ch_names])

# =========================================================================
# STEP 2: HIGIENIZAÇÃO DE NANS E DC OFFSET
# =========================================================================
data = raw.get_data()

if np.isnan(data).any():
    print("Valores nulos (NaN) detectados! Interpolando suavemente...")
    df_data = pd.DataFrame(data.T).interpolate(method='linear', limit_direction='both')
    raw._data = df_data.values.T

print("Removendo DC offset (Centralizando em zero)...")
raw.apply_function(lambda x: x - np.mean(x))

bads = []
if 'AFz' in raw.ch_names:
    bads.append('AFz')

data_limpa = raw.get_data()
for i, ch in enumerate(raw.ch_names):
    if np.var(data_limpa[i]) < 1e-12:
        if ch not in bads:
            bads.append(ch)

raw.info['bads'] = bads
print(f"Canais na quarentena: {bads}")

print("Reduzindo amostragem para 256 Hz...")
raw.resample(256.0)

raw_antes = raw.copy() # Salva cópia original para o gráfico

# =========================================================================
# STEP 3: O PIPELINE DO ARTIGO (FILTRO 0.3-70Hz + MAD + ICA)
# =========================================================================
print("\n--- INICIANDO LIMPEZA ICA ---")
print("1. Filtrando para o ICA (0.3 - 70 Hz)...")
iir_params_ica = dict(order=4, ftype='butter', output='sos')
raw_ica = raw.copy().filter(l_freq=0.3, h_freq=70.0, method='iir', iir_params=iir_params_ica, picks='eeg', verbose=False)

print("2. Calculando o Limiar MAD para ignorar picos extremos...")
data_ica = raw_ica.get_data(picks='eeg')
sfreq = raw_ica.info['sfreq']
bad_times = np.zeros(data_ica.shape[1], dtype=bool)

for i in range(data_ica.shape[0]):
    ch_data = data_ica[i, :]
    mad = np.median(np.abs(ch_data - np.median(ch_data)))
    threshold = 10.4 * mad
    bad_times |= (np.abs(ch_data) > threshold)

# Convertendo booleanos para Annotations do MNE
diffs = np.diff(bad_times.astype(int))
starts = np.where(diffs == 1)[0] + 1
ends = np.where(diffs == -1)[0] + 1
if bad_times[0]: starts = np.insert(starts, 0, 0)
if bad_times[-1]: ends = np.append(ends, len(bad_times))

onsets, durations = starts / sfreq, (ends - starts) / sfreq
if len(onsets) > 0:
    # Extraímos o carimbo de tempo exato do arquivo original
    tempo_original = raw_ica.annotations.orig_time

    # Criamos as anotações passando esse mesmo relógio
    mad_annots = mne.Annotations(onset=onsets, duration=durations, description='BAD_MAD', orig_time=tempo_original)

    # Agora o MNE vai aceitar a soma sem reclamar
    raw_ica.set_annotations(raw_ica.annotations + mad_annots)
    print(f"   -> {len(onsets)} segmentos ruidosos marcados.")

print("3. Treinando ICA (Infomax Extended, 99% Variância)...")
ica = mne.preprocessing.ICA(n_components=0.99, method='infomax', fit_params=dict(extended=True), random_state=42)
ica.fit(raw_ica, reject_by_annotation=True, verbose=False)

print("4. Detectando artefatos oculares com o MNE Nativo...")
# Buscamos um canal frontal para servir de referência de olho (EOG virtual)
canais_frontais = [ch for ch in raw.ch_names if 'F' in ch and ch not in bads]
if canais_frontais:
    canal_eog_virtual = canais_frontais[0] # Ex: Pega Fz ou Fp1
    print(f"   -> Usando {canal_eog_virtual} como referência para piscar de olhos.")
    # O MNE cruza a assinatura do ICA com a do canal frontal para achar o olho
    eog_indices, _ = ica.find_bads_eog(raw_ica, ch_name=canal_eog_virtual, verbose=False)
    ica.exclude = eog_indices
    print(f"   -> Componentes removidos: {eog_indices}")
else:
    print("   -> Nenhum canal frontal encontrado para detecção automática.")

print("5. Removendo artefatos do sinal bruto...")
raw_limpo = raw.copy()
ica.apply(raw_limpo, verbose=False)

# =========================================================================
# STEP 4: FILTRAGEM MRCP E CAR (SINAL LIMPO)
# =========================================================================
print("\n--- EXTRAINDO O MRCP ---")
print("Aplicando filtro Butterworth restrito (0.3 - 3.0 Hz)...")
iir_params_mrcp = dict(order=4, ftype='butter', output='sos')

raw_mrcp = raw_limpo.filter(l_freq=0.3, h_freq=40.0, method='iir', iir_params=iir_params_mrcp, picks='eeg', verbose=False)

print("Aplicando Common Average Reference (CAR)...")
raw_mrcp.set_eeg_reference('average', projection=False, verbose=False)

# =========================================================================
# DIAGNÓSTICO E NOVA PLOTAGEM
# =========================================================================
print("\n--- DIAGNÓSTICO DE ESCALA ---")
start_idx, stop_idx = raw_mrcp.time_as_index([10.0, 20.0])
plot_times = raw_mrcp.times[start_idx:stop_idx]

# Vamos pegar os dados PUROS primeiro, sem multiplicar por nada
data_antes_puro = raw_antes.get_data(picks=CANAIS_MOTOR, start=start_idx, stop=stop_idx)
data_depois_puro = raw_mrcp.get_data(picks=CANAIS_MOTOR, start=start_idx, stop=stop_idx)

max_puro = np.max(np.abs(data_antes_puro))
print(f"Maior valor no sinal original (entre 10 e 20s): {max_puro}")

# Lógica de auto-correção de escala:
# Se o valor já for maior que 1, o GDF já estava em microvolts!
if max_puro > 1.0:
    print("-> ATENÇÃO: O GDF parece já estar em Microvolts! Não vamos multiplicar por 1e6.")
    fator_escala = 1.0
else:
    print("-> O GDF está em Volts (Padrão MNE). Multiplicando por 1e6 para o gráfico.")
    fator_escala = 1e6

data_antes = data_antes_puro * fator_escala
data_depois = data_depois_puro * fator_escala

fig, axes = plt.subplots(3, 2, figsize=(14, 10))
axes = axes.flatten()

for i, ch_name in enumerate(CANAIS_MOTOR):
    ax = axes[i]

    if ch_name in bads:
        ax.set_title(f"Canal {ch_name} (MORTO)", color='red')
        continue

    ax.plot(plot_times, data_antes[i], color='red', alpha=0.3, label='Original')
    ax.plot(plot_times, data_depois[i], color='black', alpha=1.0, label='Filtrado (0.3-40Hz)')

    ax.set_title(f"Canal {ch_name}", fontweight='bold')
    ax.set_xlabel("Tempo (s)")
    ax.set_ylabel("Amplitude")
    ax.grid(True, linestyle='--', alpha=0.6)

    # REMOVEMOS O ax.set_ylim(-30, 30) !!
    # Agora o gráfico vai se ajustar automaticamente ao tamanho real do sinal.

    if i == 0:
        ax.legend(loc='upper right')

axes[5].axis('off')
plt.tight_layout()
plt.show()

# BÔNUS: O plot nativo do MNE (que é interativo e escala sozinho)
# Descomente a linha abaixo se quiser navegar pelo sinal inteiro deslizando a tela!
# raw_mrcp.pick(CANAIS_MOTOR).plot(duration=10.0, scalings='auto', title="MRCP Interativo")

# =========================================================================
# STEP 5: EPOCHING E REJEIÇÃO DE TENTATIVAS
# =========================================================================
print("\n--- FATIAMENTO E REJEIÇÃO DE ÉPOCAS ---")

events, event_dict = mne.events_from_annotations(raw_mrcp)

# O seu novo dicionário traduzido do artigo!
# Usamos os IDs mapeados pelo MNE (1 a 7) correspondentes aos códigos 1536 a 1542
eventos_mapeados = {
    'Elbow_Flexion': 1,
    'Elbow_Extension': 2,
    'Supination': 3,
    'Pronation': 4,
    'Hand_Close': 5,
    'Hand_Open': 6,
    'Rest': 7
}

# Limiar corrigido para a escala nativa do seu GDF (em microvolts)
reject_criteria = dict(eeg=100.0)

tmin = -1.0
tmax = 3.0

print(f"Cortando épocas ({tmin}s até {tmax}s) e aplicando rejeição de 100 µV...")
epochs = mne.Epochs(
    raw_mrcp,
    events,
    event_id=eventos_mapeados, # Agora passamos apenas os eventos que importam!
    tmin=tmin,
    tmax=tmax,
    baseline=None,
    reject=reject_criteria,
    reject_by_annotation=True,
    event_repeated='drop',
    preload=True,
    verbose=False # Deixei False para o log ficar mais limpo
)

print("\n=== RESUMO FINAL DO PRÉ-PROCESSAMENTO ===")
print(epochs)

# Mostra quantas épocas de cada movimento sobreviveram à limpeza
for evento, id_evento in eventos_mapeados.items():
    quantidade = len(epochs[evento])
    print(f"{evento}: {quantidade} trials limpos")

# =========================================================================
# STEP 6: EXPORTANDO OS DADOS LIMPOS
# =========================================================================
print("\n--- SALVANDO OS DADOS ---")

# Vamos criar caminhos para salvar na mesma pasta do seu projeto
caminho_salvar_csv = "/content/drive/MyDrive/IA753-Trabalho_Final/Data/S02_ME/S02_ME_dados_limpos.csv"
caminho_salvar_fif = "/content/drive/MyDrive/IA753-Trabalho_Final/Data/S02_ME/S02_ME_dados_limpos-epo.fif"

# ---------------------------------------------------------
# OPÇÃO 1: Salvando em CSV (Como você pediu)
# ---------------------------------------------------------
print("Convertendo para tabela Pandas e salvando CSV...")
df_epochs = epochs.to_data_frame()
df_epochs.to_csv(caminho_salvar_csv, index=False)
print(f"-> CSV salvo com sucesso em: {caminho_salvar_csv}")

# ---------------------------------------------------------
# OPÇÃO 2: Salvando no Formato Nativo do MNE (Recomendação de Ouro)
# ---------------------------------------------------------
print("Salvando no formato nativo FIF...")
epochs.save(caminho_salvar_fif, overwrite=True)
print(f"-> FIF salvo com sucesso em: {caminho_salvar_fif}")
```

```python
#-----------------------------------------------------------------------------
#    PROCESSA TODAS AS RUNS DE UM PACIENTE E SALVA EM ARQUIVOS SEPARADOS
#-----------------------------------------------------------------------------

import mne
import numpy as np
import pandas as pd

# =========================================================================
# CONFIGURAÇÕES GERAIS
# =========================================================================
CANAIS_MOTOR = ["C3", "C1", "Cz", "C2", "C4"] # <-- Adicionado aqui!

eventos_mapeados = {
    'Elbow_Flexion': 1, 'Elbow_Extension': 2, 'Supination': 3,
    'Pronation': 4, 'Hand_Close': 5, 'Hand_Open': 6, 'Rest': 7
}
reject_criteria = dict(eeg=100.0)

# =========================================================================
# LOOP DE PROCESSAMENTO EM LOTE (RUNS 1 A 10)
# =========================================================================
for run_idx in range(1, 11):

    # NOME DINÂMICO DO ARQUIVO DE ENTRADA
    arquivo = f"/content/drive/MyDrive/IA753-Trabalho_Final/Data/S02_MI/motorimagination_subject2_run{run_idx}.gdf"

    print(f"\n{'='*60}")
    print(f"PROCESSANDO A RUN {run_idx}...")
    print(f"{'='*60}")

    try:
        # =========================================================================
        # STEP 1: CARREGAMENTO E LIMPEZA DE INTRUSOS
        # =========================================================================
        print("Carregando arquivo GDF...")
        raw = mne.io.read_raw_gdf(arquivo, preload=True, verbose=False)

        canais_remover = [
            'thumb_near', 'thumb_far', 'thumb_index', 'index_near', 'index_far',
            'index_middle', 'middle_near', 'middle_far', 'middle_ring', 'ring_near',
            'ring_far', 'ring_little', 'litte_near', 'litte_far', 'thumb_palm',
            'wrist_bend', 'roll', 'pitch', 'gesture', 'handPosX', 'handPosY',
            'handPosZ', 'elbowPosX', 'elbowPosY', 'elbowPosZ', 'ShoulderAdductio',
            'ShoulderFlexionE', 'ShoulderRotation', 'Elbow', 'ProSupination',
            'Wrist', 'GripPressure'
        ]
        raw.drop_channels([ch for ch in canais_remover if ch in raw.ch_names])

        # =========================================================================
        # STEP 2: HIGIENIZAÇÃO DE NANS E DC OFFSET
        # =========================================================================
        data = raw.get_data()

        if np.isnan(data).any():
            print("Valores nulos (NaN) detectados! Interpolando suavemente...")
            df_data = pd.DataFrame(data.T).interpolate(method='linear', limit_direction='both')
            raw._data = df_data.values.T

        print("Removendo DC offset (Centralizando em zero)...")
        raw.apply_function(lambda x: x - np.mean(x))

        bads = []
        if 'AFz' in raw.ch_names:
            bads.append('AFz')

        data_limpa = raw.get_data()
        for i, ch in enumerate(raw.ch_names):
            if np.var(data_limpa[i]) < 1e-12:
                if ch not in bads:
                    bads.append(ch)

        raw.info['bads'] = bads
        print(f"Canais na quarentena: {bads}")

        print("Reduzindo amostragem para 256 Hz...")
        raw.resample(256.0)

        # =========================================================================
        # STEP 3: O PIPELINE DO ARTIGO (FILTRO 0.3-70Hz + MAD + ICA)
        # =========================================================================
        print("INICIANDO LIMPEZA ICA...")
        iir_params_ica = dict(order=4, ftype='butter', output='sos')
        raw_ica = raw.copy().filter(l_freq=0.3, h_freq=70.0, method='iir', iir_params=iir_params_ica, picks='eeg', verbose=False)

        data_ica = raw_ica.get_data(picks='eeg')
        sfreq = raw_ica.info['sfreq']
        bad_times = np.zeros(data_ica.shape[1], dtype=bool)

        for i in range(data_ica.shape[0]):
            ch_data = data_ica[i, :]
            mad = np.median(np.abs(ch_data - np.median(ch_data)))
            threshold = 10.4 * mad
            bad_times |= (np.abs(ch_data) > threshold)

        diffs = np.diff(bad_times.astype(int))
        starts = np.where(diffs == 1)[0] + 1
        ends = np.where(diffs == -1)[0] + 1
        if bad_times[0]: starts = np.insert(starts, 0, 0)
        if bad_times[-1]: ends = np.append(ends, len(bad_times))

        onsets, durations = starts / sfreq, (ends - starts) / sfreq
        if len(onsets) > 0:
            tempo_original = raw_ica.annotations.orig_time
            mad_annots = mne.Annotations(onset=onsets, duration=durations, description='BAD_MAD', orig_time=tempo_original)
            raw_ica.set_annotations(raw_ica.annotations + mad_annots)

        ica = mne.preprocessing.ICA(n_components=0.99, method='infomax', fit_params=dict(extended=True), random_state=42)
        ica.fit(raw_ica, reject_by_annotation=True, verbose=False)

        canais_frontais = [ch for ch in raw.ch_names if 'F' in ch and ch not in bads]
        if canais_frontais:
            canal_eog_virtual = canais_frontais[0]
            eog_indices, _ = ica.find_bads_eog(raw_ica, ch_name=canal_eog_virtual, verbose=False)
            ica.exclude = eog_indices

        raw_limpo = raw.copy()
        ica.apply(raw_limpo, verbose=False)

        # =========================================================================
        # STEP 4: FILTRAGEM FINAL E CAR (SINAL LIMPO)
        # =========================================================================
        print("Aplicando filtro (0.3 - 40 Hz) e CAR...")
        iir_params_mrcp = dict(order=4, ftype='butter', output='sos')
        raw_mrcp = raw_limpo.filter(l_freq=0.3, h_freq=40.0, method='iir', iir_params=iir_params_mrcp, picks='eeg', verbose=False)
        raw_mrcp.set_eeg_reference('average', projection=False, verbose=False)

        # =========================================================================
        # STEP 5: FATIAMENTO, REJEIÇÃO E SALVAMENTO
        # =========================================================================
        print("Cortando épocas (-1.0s até 3.0s) e aplicando rejeição...")
        events, _ = mne.events_from_annotations(raw_mrcp, verbose=False)

        epochs = mne.Epochs(
            raw_mrcp, events, event_id=eventos_mapeados,
            tmin=-1.0, tmax=3.0, baseline=None,
            reject=reject_criteria, reject_by_annotation=True,
            event_repeated='drop', preload=True, verbose=False
        )

        # NOME DINÂMICO DO ARQUIVO DE SAÍDA (O CHECKPOINT)
        caminho_salvar_fif = f"/content/drive/MyDrive/IA753-Trabalho_Final/Data/S02_MI/S02_MI_run{run_idx}_limpo-epo.fif"
        epochs.save(caminho_salvar_fif, overwrite=True)
        print(f"✅ Run {run_idx} concluída e salva com {len(epochs)} épocas em: {caminho_salvar_fif}")

    except FileNotFoundError:
         print(f"⚠️ Arquivo da Run {run_idx} não encontrado. Pulando para a próxima...")
    except Exception as e:
        print(f"❌ Erro inesperado ao processar a run {run_idx}: {e}")

print("\n🎉 PROCESSAMENTO EM LOTE FINALIZADO!")
```

```python
import mne
import numpy as np
import pandas as pd
import os

# =========================================================================
# CONFIGURAÇÕES GERAIS - PACIENTE SCI
# =========================================================================
CANAIS_MOTOR = ["C3", "C1", "Cz", "C2", "C4"]
reject_criteria = dict(eeg=100.0)

# O pulo do gato: Apenas as runs de movimento, ignorando Olhos e Repouso!
runs_sci = [3, 4, 5, 6, 7, 10, 11, 12, 13]

# =========================================================================
# LOOP DE PROCESSAMENTO EM LOTE (SCI)
# =========================================================================
for run_idx in runs_sci:

    # NOME DINÂMICO DO ARQUIVO DE ENTRADA (Ajuste o caminho se necessário)
    arquivo = f"/content/drive/MyDrive/IA753-Trabalho_Final/Data/P02/P02 Run {run_idx}.gdf"

    print(f"\n{'='*60}")
    print(f"PROCESSANDO A RUN {run_idx} (SCI)...")
    print(f"{'='*60}")

    try:
        # =========================================================================
        # STEP 1: CARREGAMENTO E LIMPEZA
        # =========================================================================
        raw = mne.io.read_raw_gdf(arquivo, preload=True, verbose=False)

        canais_remover = [
            'thumb_near', 'thumb_far', 'thumb_index', 'index_near', 'index_far',
            'index_middle', 'middle_near', 'middle_far', 'middle_ring', 'ring_near',
            'ring_far', 'ring_little', 'litte_near', 'litte_far', 'thumb_palm',
            'wrist_bend', 'roll', 'pitch', 'gesture', 'handPosX', 'handPosY',
            'handPosZ', 'elbowPosX', 'elbowPosY', 'elbowPosZ', 'ShoulderAdductio',
            'ShoulderFlexionE', 'ShoulderRotation', 'Elbow', 'ProSupination',
            'Wrist', 'GripPressure'
        ]
        raw.drop_channels([ch for ch in canais_remover if ch in raw.ch_names])

        # =========================================================================
        # STEP 2: HIGIENIZAÇÃO DE NANS E DC OFFSET
        # =========================================================================
        data = raw.get_data()

        if np.isnan(data).any():
            print("Interpolando NaN...")
            df_data = pd.DataFrame(data.T).interpolate(method='linear', limit_direction='both')
            raw._data = df_data.values.T

        raw.apply_function(lambda x: x - np.mean(x))

        bads = []
        if 'AFz' in raw.ch_names: bads.append('AFz')

        data_limpa = raw.get_data()
        for i, ch in enumerate(raw.ch_names):
            if np.var(data_limpa[i]) < 1e-12 and ch not in bads:
                bads.append(ch)

        raw.info['bads'] = bads
        raw.resample(256.0)

        # =========================================================================
        # STEP 3: ICA (Filtro 0.3-70Hz + MAD)
        # =========================================================================
        print("Executando ICA e remoção de ruídos musculares/oculares...")
        iir_params_ica = dict(order=4, ftype='butter', output='sos')
        raw_ica = raw.copy().filter(l_freq=0.3, h_freq=70.0, method='iir', iir_params=iir_params_ica, picks='eeg', verbose=False)

        data_ica = raw_ica.get_data(picks='eeg')
        sfreq = raw_ica.info['sfreq']
        bad_times = np.zeros(data_ica.shape[1], dtype=bool)

        for i in range(data_ica.shape[0]):
            ch_data = data_ica[i, :]
            mad = np.median(np.abs(ch_data - np.median(ch_data)))
            threshold = 10.4 * mad
            bad_times |= (np.abs(ch_data) > threshold)

        diffs = np.diff(bad_times.astype(int))
        starts = np.where(diffs == 1)[0] + 1
        ends = np.where(diffs == -1)[0] + 1
        if bad_times[0]: starts = np.insert(starts, 0, 0)
        if bad_times[-1]: ends = np.append(ends, len(bad_times))

        onsets, durations = starts / sfreq, (ends - starts) / sfreq
        if len(onsets) > 0:
            mad_annots = mne.Annotations(onset=onsets, duration=durations, description='BAD_MAD', orig_time=raw_ica.annotations.orig_time)
            raw_ica.set_annotations(raw_ica.annotations + mad_annots)

        ica = mne.preprocessing.ICA(n_components=0.99, method='infomax', fit_params=dict(extended=True), random_state=42)
        ica.fit(raw_ica, reject_by_annotation=True, verbose=False)

        canais_frontais = [ch for ch in raw.ch_names if 'F' in ch and ch not in bads]
        if canais_frontais:
            eog_indices, _ = ica.find_bads_eog(raw_ica, ch_name=canais_frontais[0], verbose=False)
            ica.exclude = eog_indices

        raw_limpo = raw.copy()
        ica.apply(raw_limpo, verbose=False)

        # =========================================================================
        # STEP 4: FILTRO FINAL MRCP (0.3 - 40 Hz) + CAR
        # =========================================================================
        iir_params_mrcp = dict(order=4, ftype='butter', output='sos')
        raw_mrcp = raw_limpo.filter(l_freq=0.3, h_freq=40.0, method='iir', iir_params=iir_params_mrcp, picks='eeg', verbose=False)
        raw_mrcp.set_eeg_reference('average', projection=False, verbose=False)

        # =========================================================================
        # STEP 5: FATIAMENTO COM O DICIONÁRIO SCI
        # =========================================================================
        events, event_dict = mne.events_from_annotations(raw_mrcp, verbose=False)

        # Mapeamento Dinâmico exclusivo para a Tabela 2 do artigo SCI
        eventos_mapeados_sci = {}
        tabela_sci = {
            '776': 'Supination',
            '777': 'Pronation',
            '779': 'Hand_Open',
            '925': 'Palmar_Grasp',
            '926': 'Lateral_Grasp'
        }

        # Se o MNE achou o código da tabela, ele cria a chave para fatiar
        for codigo_str, nome_classe in tabela_sci.items():
            if codigo_str in event_dict:
                eventos_mapeados_sci[nome_classe] = event_dict[codigo_str]

        print("Fatiando dados de movimento...")
        epochs = mne.Epochs(
            raw_mrcp, events, event_id=eventos_mapeados_sci,
            tmin=-1.0, tmax=3.0, baseline=None,
            reject=reject_criteria, reject_by_annotation=True,
            event_repeated='drop', preload=True, verbose=False
        )

        # O arquivo será salvo na mesma pasta
        caminho_salvar_fif = f"/content/drive/MyDrive/IA753-Trabalho_Final/Data/P02/P02_Run{run_idx}_limpo-epo.fif"
        epochs.save(caminho_salvar_fif, overwrite=True)
        print(f"✅ Run {run_idx} (SCI) salva com {len(epochs)} épocas limpas!")

    except FileNotFoundError:
         print(f"⚠️ Arquivo da Run {run_idx} não encontrado. Verifique o caminho.")
    except Exception as e:
        print(f"❌ Erro ao processar a run {run_idx}: {e}")

print("\n🎉 PROCESSAMENTO EM LOTE DO PACIENTE SCI FINALIZADO!")
```

```python
import mne
import matplotlib.pyplot as plt

# =========================================================================
# 1. CARREGAR OS TRÊS ARQUIVOS MESTRES
# =========================================================================
print("Carregando as três bases de dados...")
caminho_me = "/content/drive/MyDrive/IA753-Trabalho_Final/Data/S02_ME/S02_ME_Mestre_Limpo-epo.fif"
caminho_mi = "/content/drive/MyDrive/IA753-Trabalho_Final/Data/S02_MI/S02_MI_Mestre_Limpo-epo.fif"
caminho_sci = "/content/drive/MyDrive/IA753-Trabalho_Final/Data/P02/P02_SCI_Mestre_Limpo-epo.fif"

epochs_me = mne.read_epochs(caminho_me, preload=True, verbose=False)
epochs_mi = mne.read_epochs(caminho_mi, preload=True, verbose=False)
epochs_sci = mne.read_epochs(caminho_sci, preload=True, verbose=False)

# Colocando a touca em todos para manter a coerência espacial
montage = mne.channels.make_standard_montage('standard_1005')
epochs_me.set_montage(montage, match_case=False, on_missing='ignore')
epochs_mi.set_montage(montage, match_case=False, on_missing='ignore')
epochs_sci.set_montage(montage, match_case=False, on_missing='ignore')

# =========================================================================
# 2. O AJUSTE FINO CIENTÍFICO (ESCALA E BASELINE)
# =========================================================================
print("Aplicando correção de escala e Baseline (-1.0s a 0.0s)...")

for epocas in [epochs_me, epochs_mi, epochs_sci]:
    # 1. Corrige o Bug do Milhão
    epocas.apply_function(lambda x: x / 1e6)
    # 2. Força todas as linhas a partirem do zero absoluto antes do movimento
    epocas.apply_baseline(baseline=(-1.0, 0.0), verbose=False)

canais_motor = ["C3", "C1", "Cz", "C2", "C4"]

# =========================================================================
# 3. PREPARANDO OS DADOS PARA A COMPARAÇÃO (COM SOMBRAS DE CONFIANÇA)
# =========================================================================
# Extraímos a classe 'Hand_Open' comum a todos e fatiamos os canais motores
condicoes = {
    'Saudável (ME)': list(epochs_me['Hand_Open'].copy().pick(canais_motor).iter_evoked()),
    'Saudável (MI)': list(epochs_mi['Hand_Open'].copy().pick(canais_motor).iter_evoked()),
    'Lesão Medular (SCI)': list(epochs_sci['Hand_Open'].copy().pick(canais_motor).iter_evoked())
}

# Cores e estilos distintos para bater o olho e entender
cores = {
    'Saudável (ME)': 'red',
    'Saudável (MI)': 'blue',
    'Lesão Medular (SCI)': 'green'
}
estilos = {
    'Saudável (ME)': {"linewidth": 2.5},
    'Saudável (MI)': {"linewidth": 2, "linestyle": "--"},
    'Lesão Medular (SCI)': {"linewidth": 2.5, "linestyle": "-."}
}

# =========================================================================
# 4. PLOTAGEM A: MÉDIA DA REGIÃO MOTORA (O GRÁFICO PRINCIPAL)
# =========================================================================
print("Gerando gráficos...")

fig_roi = mne.viz.plot_compare_evokeds(
    condicoes,
    combine='mean', # Média dos 5 canais do córtex motor
    colors=cores,
    styles=estilos,
    title="Análise Comparativa do MRCP: ME x MI x SCI (Hand Open)",
    show_sensors=True,
    truncate_yaxis='auto',
    show=False # Segura pra mostrar tudo junto no final
)

# =========================================================================
# 5. PLOTAGEM B: COMPARAÇÃO ISOLADA DO HEMISFÉRIO CONTRALATERAL (C3)
# =========================================================================
# Como é a mão direita, o canal C3 (lado esquerdo) é o principal ator.
fig_c3 = mne.viz.plot_compare_evokeds(
    condicoes,
    picks="C3", # Isolando o alvo exato
    colors=cores,
    styles=estilos,
    title="Análise Focada: Canal C3 (Córtex Motor Primário Contralateral)",
    show_sensors=False,
    truncate_yaxis='auto',
    show=False
)

plt.show()
```
